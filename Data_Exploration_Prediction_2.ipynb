## Parameter tuning using cross-validation
Select the best performing n_estimators. Chose 900 n_estimators as it gives best score

```python
# estimators = [500,1000,1500,2000]
# for n_est in estimators:
#     res = cross_val_score(RandomForestClassifier(n_estimators = n_est, random_state = 42),features,labels)
#     print(res.mean())
```

```python
estimators = list(range(500,2000,100))
for n_est in estimators:
    res = cross_val_score(RandomForestClassifier(n_estimators = n_est, random_state = 42),features,labels)
    print(n_est, res.mean())
```
